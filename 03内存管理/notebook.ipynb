### new delete 使用
内置类型：  
```cpp
void Test()
{
    // 动态申请一个int类型的空间
    int* ptr4 = new int;
    // 动态申请一个int类型的空间并初始化为10
    int* ptr5 = new int(10);
    // 动态申请10个int类型的空间
    int* ptr6 = new int[3];
    delete ptr4;
    delete ptr5;
    delete[] ptr6;
}
```
操作自定义类型：  
`自动调用构造函数和析构函数，且如果申请失败则抛异常`
```cpp

```

`重点：`  
new = operator new(开空间，等于malloc，在这里抛异常) + 调用构造函数  
delete = 调用析构函数 + operator delete(相当于free)  
new[] = operator new[] + 调用n次构造函数  
delete[] = 调用n次析构函数 + operator delete[]  

注：上述调用构造和析构前提是自定义类型

`new[]和delete[]必须要匹配使用`  
比如：某个类A中显式实现了构造函数和析构函数
```cpp
A* pa = new A[10];
delete pa;
```
程序可能崩溃，因为new A[10]的时候编译器会在前面生成一个Cookie记录new的个数  
而在delete的时候，由于不是delete[]，所以会只调用一次析构（这还不是致命的）  
致命的点在于最后free的时候，会从pa开始free，但是pa位置并不是合法的起始位置，而是偏移后的非法地址，所以会崩溃  
但如果是delete[],就会从Cookie那个位置开始free，就是正常的
