### 一、string的遍历：四种迭代器  
+ iterator
+ const_iterator
+ reverse_iterator
+ const_reverse_iterator

```cpp
string s("hello world");

auto it = s.begin();
while(it != s.end())
{
    cout << *s;
    ++it;
}

auto rit = s.rbegin();
while(rit != s.rend())
{
    cout << *s;
    ++rit;
}
```
`注意这里的rit倒序遍历的时候也要用++`

范围for：  
```cpp
string s("hello world");
for(auto ch : s)
{
    cout << ch;
}
```

### 二、容量相关接口  
+ size() 或 length()，完全一样  
+ capacity()接口可以查看实际容量
+ reserve(100)可以预留100个空间，防止频繁扩容，是否缩容不一定，但一定不会修改字符串内容和长度
+ clear()：清除数据，不清容量
+ empty()：判空

### 三、修改相关操作
+ operator[] 其实有两个版本，一个正常，一个const版本，如果越界，断言抛异常
+ +=（最常用）
